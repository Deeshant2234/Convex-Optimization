```python
#CVXPY VARIABLES AND PARAMETERS
N=50
a=cp.Variable((N,1),name='a')
b=cp.Variable(name='b')

#defining a nonnegative parameter
lamd=cp.Parameter(nonneg=True,name='lamdba')

#objective
f1=cp.norm(a,2)
f2=cp.norm(a,1)
obj=f1+lamd*f2

#Constraints
constr=[]
constr+=[a.T@X[:,i]-b>=1 for i in range(N)]
constr+=[a.T@Y[:,i]-b<=-1 for i in range(N)]
#objective
objective=cp.Minimize(obj)

#Constructing the trade-off
slab_thickness=np.array([])
card_a=np.array([])



lamd_values=np.logspace(-5,5,num=100)
a_value10=[]
for value in lamd_values:
  lamd.value=value
  #problem
  prob=cp.Problem(objective,constr)
  prob.solve()
  
  k=0.0001
  count=0
  
  for a_i in a.value:
    if a_i>=k or a_i<=-k :
      count+=1

  card_a=np.append(card_a,count)
  #count=card_a(a.value)

  slabThickness=find_slabThickness(a)
  slab_thickness=np.append(slab_thickness,slabThickness.value)

  if count==10: #when cardinality=10
    a_value10.append(a.value)


#trade-off between f1 and f2
f1_value=np.array([])
f2_value=np.array([])
for a in a.value:
  f1=cp.norm(a,2)
  f2=cp.norm(a,1)
  f1_value=np.append(f1_value,f1.value)
  f2_value=np.append(f2_value,f2.value)

plt.figure()
plt.plot(f1_value,f2_value)
plt.title("Trade-off between $f_1=cp.norm(a,2)$ and $f_2=cp.norm(a,1)$")
plt.grid(True)
plt.axis(True)
plt.xlabel("$f_1=cp.norm(a,2)$")
plt.ylabel("$f_2=cp.norm(a,1)$")

a_value10final=a_value10[0] # taking any a value for which cardinality =10

a_value10final=list(a_value10final)

#selecting indexes for which a value is non zero
indices=np.array([])
for a in a_value10final:
  if a>=k or a<=-k :
    k1=a_value10final.index(a)
    indices=np.append(indices,k1)

#print("CVXPY problem is",prob)
print("Non zero indices when number of features =10")
print(indices)
print("slab thickness is",slab_thickness) 
print("cardinality of a:",card_a)
plt.figure()
plt.plot(card_a,slab_thickness,color='r')
plt.grid(True)
plt.axis(True)
plt.title("Slab thickness versus cardinality of a")
plt.ylabel("SLAB THICKNESS")
plt.xlabel("Cardinality of a")
```
