```python
import numpy as np
import cvxpy as cp
import matplotlib.pyplot as plt
f_0=1
f_1=0
f_2=2
f_3=1

#For f1(x)
x=np.linspace(0,3,100)
optval_f1x=np.array([])
for i in x:
  m=cp.Variable(name='m')
  c=cp.Variable(name='c')
  f1_x=m*i+c
  obj=cp.Minimize(f1_x)
  constr=[f_0<=c,f_1<=c+m,f_2<=2*m+c,f_3<=3*m+c]
  prob=cp.Problem(obj,constr)  
  optval_f1x=np.append(optval_f1x,prob.solve())

print("CVXPY problem for f_1(x)",prob)
plt.plot(x,optval_f1x,'o')
plt.xlabel("x-values point")
plt.ylabel("$f_1(x)$")
plt.title("f1_x curve using cvxpy ")
```

```python
#For f2(x)
x=np.linspace(0,3,100)
N=4
optval_f2x=np.array([])
for i in x:
  alpha=cp.Variable(4,name="alpha")
  c=cp.Variable()
  f2_x=alpha[0]*f_0+alpha[1]*f_1+alpha[2]*f_2+alpha[3]*f_3
  obj=cp.Maximize(f2_x)
  sum=alpha[1]+2*alpha[2]+3*alpha[3]
  sum_alpha=alpha[0]+alpha[1]+alpha[2]+alpha[3]
  constr=[sum==i,sum_alpha==1,alpha>=0]
  prob=cp.Problem(obj,constr)  
  optval_f2x=np.append(optval_f2x,prob.solve())

print("CVXPY problem for f_2(x)",prob)
plt.plot(x,optval_f2x,'o')
plt.xlabel("x-values point")
plt.ylabel("$f_2(x)$")
plt.title("f2_x curve using cvxpy ")
```

```python
#For f3(x)
x=np.linspace(0,3,100)
N=4
optval_f3x=np.array([])
for i in x:
  alpha=cp.Variable(4,name='alpha')
  f3_x=alpha[0]*f_0+alpha[1]*f_1+alpha[2]*f_2+alpha[3]*f_3
  obj=cp.Minimize(f3_x)
  sum=alpha[1]+2*alpha[2]+3*alpha[3]
  sum_alpha=alpha[0]+alpha[1]+alpha[2]+alpha[3]
  constr=[sum==i,sum_alpha==1,alpha>=0]
  prob=cp.Problem(obj,constr)  
  optval_f3x=np.append(optval_f3x,prob.solve())

print("CVXPY problem for f_3(x)",prob)
plt.plot(x,optval_f3x,'o')
plt.xlabel("x-values point")
plt.ylabel("$f_3(x)$")
plt.title("f3_x curve using cvxpy ")
```
